```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error, r2_score
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
import datetime

# 1. Fetch stock data
def get_stock_data(symbol, start='2015-01-01', end='2024-12-31'):
    try:
        data = yf.download(symbol, start=start, end=end)
        if data.empty:
            raise ValueError("No data fetched. Check ticker symbol or date range.")
        return data[['Close']], data.index
    except Exception as e:
        print(f"Error fetching data: {e}")
        return None, None

# 2. Preprocess data
def preprocess_data(df, sequence_length=60):
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(df)

    X, y = [], []
    for i in range(sequence_length, len(scaled_data)):
        X.append(scaled_data[i-sequence_length:i, 0])
        y.append(scaled_data[i, 0])

    X, y = np.array(X), np.array(y)
    X = X.reshape((X.shape[0], X.shape[1], 1))  # LSTM expects 3D input
    return X, y, scaler, scaled_data

# 3. Build the LSTM model
def build_model(input_shape):
    model = Sequential()
    model.add(LSTM(100, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))  # Reduced dropout for better learning
    model.add(LSTM(50, return_sequences=False))  # Simplified second LSTM layer
    model.add(Dropout(0.2))
    model.add(Dense(25))  # Added intermediate dense layer
    model.add(Dense(1))   # Final prediction layer
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# 4. Predict future prices
def predict_future_prices(model, recent_data, days, scaler):
    predictions = []
    input_seq = recent_data[-60:].reshape(1, 60, 1)

    for _ in range(days):
        next_price = model.predict(input_seq, verbose=0)[0][0]
        predictions.append(next_price)
        input_seq = np.roll(input_seq, -1, axis=1)  # Shift sequence
        input_seq[0, -1, 0] = next_price  # Add new prediction

    return scaler.inverse_transform(np.array(predictions).reshape(-1, 1))

# 5. Calculate accuracy metrics
def calculate_accuracy(y_true, y_pred):
    mape = mean_absolute_percentage_error(y_true, y_pred) * 100  # In percentage
    r2 = r2_score(y_true, y_pred)
    return mape, r2

# 6. Plot predictions including future
def predict_and_plot(model, X_test, y_test, scaler, recent_data, future_days, dates_test, df_dates):
    # Predict past (test)
    predictions = model.predict(X_test, verbose=0)
    predictions = scaler.inverse_transform(predictions)
    y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))

    # Calculate accuracy
    mape, r2 = calculate_accuracy(y_test_inv, predictions)
    print(f"Test Set MAPE: {mape:.2f}%")
    print(f"Test Set R² Score: {r2:.4f}")

    # Predict future
    future_preds = predict_future_prices(model, recent_data, future_days, scaler)
    
    # Generate future dates
    last_date = df_dates[-1]
    future_dates = [last_date + datetime.timedelta(days=x) for x in range(1, future_days + 1)]

    # Plot actual vs predicted
    plt.figure(figsize=(14, 7))
    
    # Plot actual test data
    plt.plot(dates_test, y_test_inv, label='Actual Price', color='blue')
    plt.plot(dates_test, predictions, label='Predicted Price', color='orange')
    
    # Plot future predictions
    plt.plot(future_dates, future_preds, label='Future Prediction', color='green')
    
    plt.title('Stock Price Prediction with 12-Month Forecast')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# 7. Main pipeline
def main():
    company = input("Enter company ticker symbol (e.g. AAPL, GOOGL, MSFT): ")
    df, df_dates = get_stock_data(company)
    
    if df is None or df.empty:
        print("Invalid ticker or no data found.")
        return

    if len(df) < 100:  # Ensure enough data
        print("Not enough data for meaningful predictions.")
        return

    sequence_length = 60
    X, y, scaler, scaled_data = preprocess_data(df, sequence_length)

    # Split data (no shuffling for time-series)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, shuffle=False
    )

    # Get corresponding dates for test set
    test_start_idx = len(df) - len(y_test) - sequence_length
    dates_test = df_dates[test_start_idx:test_start_idx + len(y_test)]

    # Build and train model
    model = build_model((X_train.shape[1], 1))
    model.fit(X_train, y_train, epochs=30, batch_size=32, verbose=1, validation_split=0.1)

    # Predict ~12 months ahead (50 trading days ≈ 2.5 months for simplicity)
    future_days = 50
    predict_and_plot(model, X_test, y_test, scaler, scaled_data, future_days, dates_test, df_dates)

# Run the app
if __name__ == "__main__":
    main()
```

```python
import yfinance as yf
import time
from datetime import datetime
import pytz
import matplotlib.pyplot as plt

# Set dark theme for plotting
plt.style.use('dark_background')

# Check if market is open (9:30 AM to 4:00 PM Eastern, Monday-Friday)
def is_market_open():
    now = datetime.now(pytz.timezone('US/Eastern'))
    hour = now.hour
    minute = now.minute
    weekday = now.weekday()
    
    # Market is open if it's Mon-Fri and between 9:30 AM and 4:00 PM
    if weekday < 5:  # 0-4 is Monday-Friday
        if (hour == 9 and minute >= 30) or (10 <= hour < 16):
            return True
    return False

# Get current stock price
def get_stock_price(ticker):
    try:
        stock = yf.Ticker(ticker)
        price = stock.info.get('regularMarketPrice')  # Current price
        return price
    except:
        return None

# Plot historical data when market is closed
def plot_historical_data(ticker):
    end_date = datetime.now().strftime("%Y-%m-%d")
    data = yf.download(ticker, start="2020-01-01", end=end_date)
    
    plt.figure(figsize=(10, 5))
    plt.plot(data.index, data['Close'], label=f'{ticker} Close Price', color='cyan')
    plt.title(f'{ticker} Closing Price (2020 - Present)', color='white')
    plt.xlabel('Date', color='white')
    plt.ylabel('Price (USD)', color='white')
    plt.legend()
    plt.grid(True, color='gray', linestyle='--', alpha=0.5)
    plt.tick_params(axis='x', colors='white')
    plt.tick_params(axis='y', colors='white')
    plt.tight_layout()
    plt.show()

# Main function to monitor stock and update plot
def watch_stock(ticker):
    last_price = None
    times = []
    prices = []
    
    # Initialize plot for real-time updates
    plt.ion()  # Turn on interactive mode
    fig, ax = plt.subplots(figsize=(10, 5))
    line, = ax.plot([], [], label=f'{ticker} Real-Time Price', color='cyan')
    ax.set_title(f'{ticker} Real-Time Price', color='white')
    ax.set_xlabel('Time', color='white')
    ax.set_ylabel('Price (USD)', color='white')
    ax.legend()
    ax.grid(True, color='gray', linestyle='--', alpha=0.5)
    ax.tick_params(axis='x', colors='white')
    ax.tick_params(axis='y', colors='white')
    
    print(f"Watching {ticker} price... (Press Ctrl+C to stop)")

    while True:
        if is_market_open():
            price = get_stock_price(ticker)
            if price is not None:
                now = datetime.now().strftime("%H:%M:%S")
                if last_price is None:
                    print(f"{now} | {ticker}: ${price:.2f}")
                else:
                    change = price - last_price
                    print(f"{now} | {ticker}: ${price:.2f} (Change: ${change:.2f})")
                last_price = price
                
                # Update real-time data
                times.append(datetime.now())
                prices.append(price)
                
                # Keep only the last 100 points to avoid clutter
                if len(times) > 100:
                    times.pop(0)
                    prices.pop(0)
                
                # Update plot
                line.set_xdata([t.strftime("%H:%M:%S") for t in times])
                line.set_ydata(prices)
                ax.relim()
                ax.autoscale_view()
                ax.tick_params(axis='x', rotation=45)
                fig.canvas.draw()
                fig.canvas.flush_events()
            else:
                print("Can't get price, trying again...")
        else:
            print("Market is closed, displaying historical data...")
            plt.close(fig)  # Close real-time plot
            plot_historical_data(ticker)
            time.sleep(60)  # Wait 1 minute if market is closed
            return  # Exit after showing historical plot
        
        time.sleep(10)  # Wait 10 seconds before next check

# Run the program
ticker = "AAPL"  # Example: Apple stock
try:
    watch_stock(ticker)
except KeyboardInterrupt:
    print("\nStopped watching.")
    plt.close('all')
```

```python
import yfinance as yf
import json
from datetime import datetime
from dateutil.parser import parse

# Validate date input
def validate_date(date_str):
    try:
        parsed_date = parse(date_str, fuzzy=False)
        return parsed_date.strftime("%Y-%m-%d")
    except:
        return None

# Get historical stock data
def get_historical_data(ticker, start_date, end_date):
    try:
        stock = yf.Ticker(ticker)
        
        # Fetch historical data
        history = stock.historyался

        # Convert history to list of dictionaries
        history_data = []
        for date, row in history.iterrows():
            history_data.append({
                'Date': date.strftime("%Y-%m-%d"),
                'Open': row['Open'],
                'High': row['High'],
                'Low': row['Low'],
                'Close': row['Close'],
                'Volume': row['Volume'],
                'Dividends': row['Dividends'],
                'Stock Splits': row['Stock Splits']
            })
        
        # Include ticker in the data
        data = {
            'Ticker': ticker,
            'Historical Data': history_data
        }
        return data
    except:
        return None

# Save data to JSON file
def save_to_json_file(data, ticker, start_date, end_date):
    if not data:
        return False
    filename = f"{ticker}_history_{start_date}_{end_date}.json"
    try:
        with open(filename, 'w') as file:
            json.dump(data, file, indent=4)
        return filename
    except:
        return False

# Main program
print("Enter stock ticker (e.g., RELIANCE.NS for Reliance Industries):")
ticker = input().strip()

print("Enter start date (YYYY-MM-DD, e.g., 2024-01-01):")
start_date = validate_date(input().strip())

print("Enter end date (YYYY-MM-DD, e.g., 2025-04-29):")
end_date = validate_date(input().strip())

# Validate inputs
if not start_date or not end_date:
    print("Error: Invalid date format. Use YYYY-MM-DD.")
elif start_date > end_date:
    print("Error: Start date must be before end date.")
else:
    # Get historical data
    historical_data = get_historical_data(ticker, start_date, end_date)
    
    # Save and display result
    if historical_data:
        filename = save_to_json_file(historical_data, ticker, start_date, end_date)
        if filename:
            print(f"Historical data saved to {filename}")
            print("\nData Preview:")
            print("-------------")
            print(f"Ticker: {historical_data['Ticker']}")
            print(f"Historical Data (first 2 and last 2 entries):")
            history = historical_data['Historical Data']
            for entry in history[:2] + history[-2:]:
                print(f"  {entry['Date']}: Open={entry['Open']:.2f}, Close={entry['Close']:.2f}, Volume={entry['Volume']:,}")
        else:
            print("Error: Could not save data to JSON file.")
    else:
        print(f"Error: Could not fetch data for {ticker}. Check ticker, dates, or internet.")
```

# More Feature Like news sentimental anakysisi

```python
import requests
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

# Download VADER lexicon
nltk.download('vader_lexicon', quiet=True)

# Replace this with your actual NewsAPI key
API_KEY = '597d986c60374539aeaf83ca1da405f6'

def fetch_news(ticker, api_key=API_KEY):
    """
    Fetch top 5 recent news headlines using NewsAPI for a company or stock ticker.
    """
    url = f"https://newsapi.org/v2/everything"
    params = {
        'q': ticker,
        'sortBy': 'publishedAt',
        'language': 'en',
        'pageSize': 5,
        'apiKey': api_key
    }

    try:
        response = requests.get(url, params=params)
        response.raise_for_status()
        articles = response.json().get('articles', [])
        
        headlines = [article['title'] for article in articles if 'title' in article]
        
        if not headlines:
            print(f"No headlines found for '{ticker}'.")
        else:
            print(f"Fetched {len(headlines)} headlines for '{ticker}':")
            for i, headline in enumerate(headlines, 1):
                print(f"  {i}. {headline}")
        return headlines
    except Exception as e:
        print(f"Error fetching news: {e}")
        return []

def analyze_sentiment(headlines):
    """
    Analyze sentiment of headlines using VADER.
    """
    if not headlines:
        return []

    sid = SentimentIntensityAnalyzer()
    results = []
    for headline in headlines:
        score = sid.polarity_scores(headline)
        results.append((headline, score['compound']))
        print(f"Sentiment for '{headline[:50]}...': {score['compound']:.3f}")
    return results

def plot_sentiments(results, ticker):
    """
    Plot sentiment results in a horizontal bar chart.
    """
    if not results:
        print("No sentiment data to plot.")
        return

    headlines = [h[:50] + "..." if len(h) > 50 else h for h, _ in results]
    scores = [s for _, s in results]
    colors = ['green' if s > 0.05 else 'red' if s < -0.05 else 'gray' for s in scores]

    plt.figure(figsize=(10, max(6, len(results) * 0.5)))
    bars = plt.barh(headlines, scores, color=colors)

    plt.title(f'News Sentiment for {ticker}')
    plt.xlabel('Sentiment Score')
    plt.ylabel('Headlines')
    plt.grid(True, axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()

    for bar in bars:
        width = bar.get_width()
        label_x = width + 0.02 if width > 0 else width - 0.02
        plt.text(label_x, bar.get_y() + bar.get_height()/2, f'{width:.3f}',
                 ha='left' if width > 0 else 'right', va='center')

    try:
        plt.savefig('sentiment_plot.png', bbox_inches='tight', dpi=100)
        print("Sentiment plot saved as 'sentiment_plot.png'.")
    except Exception as e:
        print(f"Error saving plot: {e}")
    finally:
        plt.close()

def main():
    ticker = "Reliance Industries"
    headlines = fetch_news(ticker)
    sentiments = analyze_sentiment(headlines)
    plot_sentiments(sentiments, ticker)

if __name__ == "__main__":
    main()

```
